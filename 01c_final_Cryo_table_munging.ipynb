

# NEW Cryopreservation data

In October 2022 we sorted 2 additional plates that had been cryopreserved 
1. A 4 year cryoprseervation from October 30th 2018 and 
1. A ~3.5 year cryopreservation from the mesopelagic JdF samples

These samples went through the new SCGC assembly process and were copied over and added to this analysis on November 29th 2022. The cells below import the assembly and FACS data for these new cells creates a new df and then combine it with the old files. This also generates new dotplots for these samples.

```python
FACS_files=glob.glob('/mnt/scgc/simon/microg2p/analyses/Cryopreservation_analysis/tbls/*-data-export.csv')

DF_FACS=pd.DataFrame()
for file in FACS_files:
    tdf=pd.read_csv(file)
    base=os.path.basename(file)
    plate=base[:6]
    print(plate)
    tdf['Plate']=plate
    tdf['name']=tdf['Plate']+'-'+tdf['well']
    tdf=tdf.drop(columns={'well'})
    tdf=tdf.rename(columns={'repiration_rate_amoles_per_cell_per_h':'Predicted_respiration_rate_fmol_O2perhr'})
    
    DF_FACS=pd.concat([DF_FACS, tdf])

DF_FACS

```

```python
Assembly_files=glob.glob('/mnt/scgc/simon/microg2p/analyses/Cryopreservation_analysis/tbls/22*_assembly_stats.csv')

DF_assembly=pd.DataFrame()
for file in Assembly_files:
    tdf=pd.read_csv(file)
    tdf=tdf.rename(columns={'well':'name'})
    DF_assembly=pd.concat([DF_assembly, tdf])

DF_assembly[['domain', 'phyla', 'class', 'order', 'family', 'genus', 'species']]=DF_assembly['GTDB_classification'].str.split(';', expand=True)
DF_assembly=DF_assembly.rename(columns={'GTDB_classification':'GTDBtk_classification'})
DF_assembly['GTDBtk_classification'].replace('0', 'Unclassified', inplace=True)


DF_assembly['phyla']=DF_assembly['phyla'].str[3:]
DF_assembly['class']=DF_assembly['class'].str[3:]
DF_assembly['order']=DF_assembly['order'].str[3:]
DF_assembly['family']=DF_assembly['family'].str[3:]
DF_assembly['genus']=DF_assembly['genus'].str[3:]
DF_assembly['species']=DF_assembly['species'].str[3:]
DF_assembly


# DF_assembly['GTDB_classification']='Unclassified'
DF_assembly.loc[DF_assembly['phyla'] != 'None', 'GTDB_classification'] = DF_assembly['phyla']
DF_assembly.loc[DF_assembly['class'] != 'None', 'GTDB_classification'] = DF_assembly['class']
DF_assembly.loc[DF_assembly['order'] != 'None', 'GTDB_classification'] = DF_assembly['order']
DF_assembly.loc[DF_assembly['family'] != 'None', 'GTDB_classification'] = DF_assembly['family']
DF_assembly.loc[DF_assembly['genus'] != 'None', 'GTDB_classification'] = DF_assembly['genus']

DF_assembly.loc[DF_assembly['GTDBtk_classification'] == 'Unclassified', 'GTDB_classification'] = 'Unclassified'


DF_assembly['GTDB_classification'].value_counts()
```

```python
DF_new_SAGs=DF_assembly.merge(DF_FACS, on='name', how='left')
DF_new_SAGs['GTDBtk_classification']=DF_new_SAGs['GTDBtk_classification'].fillna('Unclassified')
DF_new_SAGs['Predicted_respiration_rate_fmol_O2perhr']=DF_new_SAGs['Predicted_respiration_rate_fmol_O2perhr']/1000
DF_new_SAGs.GTDBtk_classification
```

```python
cryo_data=pd.concat([cryo_data, SAG_data, JdF_data, DF_new_SAGs], sort=True)
cryo_data.columns
```

```python
freeze={
    'Plate':['AH-770', 'AM-586', 'AH-545', 'AH-580', 'AM-379', 'AM-582'],
    'Freeze_duration':['84 days', '1230 days', 'Fresh', '2 days', '1064 days', '1435 days']}
DF_freeze = pd.DataFrame(freeze)

DF_freeze
```

```python
cols=['name', 'GTDB_classification', 'Predicted_respiration_rate_fmol_O2perhr', 'FACS_date', 
      'Plate', 'GTDBtk_classification']

heatmap_data=cryo_data[cols].copy()
heatmap_data['GTDBtk_classification']=heatmap_data['GTDBtk_classification'].fillna('Unclassified')
heatmap_data['GTDBtk_classification']=heatmap_data['GTDBtk_classification'].replace('0', 'Unclassified')
heatmap_data['GTDB_classification']=heatmap_data['GTDB_classification'].replace('', 'Unclassified')

# df['Feature3'] = df['dic'].apply(lambda x: x.get('Feature3'))
heatmap_data=heatmap_data.merge(DF_freeze, on='Plate', how='left')


Meso_data=heatmap_data[heatmap_data['Plate'].isin(['AH-770', 'AM-586'])]
Meso_data.to_csv('tbls/meso_data.csv')
Surf_data=heatmap_data[heatmap_data['Plate'].isin(['AH-545', 'AH-580', 'AM-379', 'AM-582'])]
Surf_data.to_csv('tbls/Surf_data.csv')

heatmap_data.to_csv('tbls/data_for_heatmap_analysis.csv')
heatmap_data
```

```python



# ADD the section to generate dotplots but make sure that I modify the values so that I am using the same colors and albels as the rest of the paper.
Plates=heatmap_data.Plate.unique()

    
cols=['name', 'GTDBtk_classification']
dotplot=heatmap_data[cols].copy()

# I am doing this one step at a time because if I loop it I am worried that I will lose the 
# ability to control the order in which each lineage is tested. Eg Bacteria will be last and all the 
dotplot['dot_classification']=np.where(dotplot.GTDBtk_classification.str.contains('a'), 'Other', '')
dotplot['dot_classification']=np.where(dotplot.GTDBtk_classification.str.contains('Alphaproteobacteria'), 'Alphaproteobacteria', dotplot['dot_classification'])
dotplot['dot_classification']=np.where(dotplot.GTDBtk_classification.str.contains('Planctomycetes'), 'Planctomycetes', dotplot['dot_classification'])
dotplot['dot_classification']=np.where(dotplot.GTDBtk_classification.str.contains('Gammaproteobacteria'), 'Gammaproteobacteria', dotplot['dot_classification'])
dotplot['dot_classification']=np.where(dotplot.GTDBtk_classification.str.contains('Poseidoniia'), 'Poseidoniia', dotplot['dot_classification'])
dotplot['dot_classification']=np.where(dotplot.GTDBtk_classification.str.contains('Nitrososphaeria'), 'Nitrososphaeria', dotplot['dot_classification'])
dotplot['dot_classification']=np.where(dotplot.GTDBtk_classification.str.contains('Verrucomicrobiae'), 'Verrucomicrobiae', dotplot['dot_classification'])
dotplot['dot_classification']=np.where(dotplot.GTDBtk_classification.str.contains('Acidimicrobiia'), 'Acidimicrobiia', dotplot['dot_classification'])
dotplot['dot_classification']=np.where(dotplot.GTDBtk_classification.str.contains('Actinomycetia'), 'Actinomycetia', dotplot['dot_classification'])
dotplot['dot_classification']=np.where(dotplot.GTDBtk_classification.str.contains('Marinisomatia'), 'Marinisomatia', dotplot['dot_classification'])
dotplot['dot_classification']=np.where(dotplot.GTDBtk_classification.str.contains('Kapabacteria'), 'Kapabacteria', dotplot['dot_classification'])
dotplot['dot_classification']=np.where(dotplot.GTDBtk_classification.str.contains('Bacteroidia'), 'Bacteroidia', dotplot['dot_classification'])

dotplot['dot_classification']=np.where(dotplot.GTDBtk_classification.str.contains('Cyanobacteriia'), 'Cyanobacteriia', dotplot['dot_classification'])
dotplot['dot_classification']=np.where(dotplot.GTDBtk_classification.str.contains('UBA1135'), 'UBA1135', dotplot['dot_classification'])
dotplot['dot_classification']=np.where(dotplot.GTDBtk_classification.str.contains('Unclassified'), 'Other', dotplot['dot_classification'])

outdir='./dotplot_tbls/'
if not os.path.exists(outdir):
    os.mkdir(outdir)
    
! rm dotplot_tbls/A*.csv

for p in Plates:
    tdf=dotplot[dotplot.name.str.contains(p, na=False)]
    tdf=tdf.drop(columns=['GTDBtk_classification'])
    tdf.to_csv(outdir+ p+'_dotplot.csv', index=False, header=False)
    

```
